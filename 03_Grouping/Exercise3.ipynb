### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users= pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|')
users
```

### Step 4. Discover what is the mean age per occupation

```python
np.round(users.groupby('occupation')['age'].agg([np.mean]),2)
#alternate way: np.round(users.groupby('occupation')['age'].mean(),1)
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
gender_count = users.groupby(['occupation', 'gender'])['gender'].agg({'Age': 'count'})
gender_count
#continue in next inline..
```

```python
#percentage ratio of gender. Value is rounded for ratio readability
gender_ratio = np.round(gender_count.groupby(level=0).apply(lambda x:100 * x / float(x.sum())) ,0)
gender_ratio
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby('occupation')['age'].agg({'min': np.min, 'max':np.max})
#alternare way: users.groupby('occupation')['age'].describe()
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
np.round(users.groupby(['occupation', 'gender'])['age'].agg({'mean age': np.mean}),2)
```

### Step 8.  For each occupation present the percentage of women and men

```python
np.round(gender_count.groupby(level=0).apply(lambda x:100 * x / float(x.sum())) ,2)
```
